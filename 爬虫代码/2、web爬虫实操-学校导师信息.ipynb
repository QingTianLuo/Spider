```python
URL = 'http://ds.zufe.edu.cn/szdw2/zyjs1.htm'#整个网页链接
URL_PREF="http://ds.zufe.edu.cn/info/1092/"#每个老师的个人链接
```

```python
res=requests.get(URL,headers=headers)
res.encoding="utf-8"
html_source=res.text
status_code=res.status_code
print(status_code)
#print(html_source)
soup=BeautifulSoup(html_source,"html.parser")
```

```python
names=[]
introduce_url=[]
department=[]
for i in range(len(soup.findAll("div",class_="zyjs-jjtj"))):
    for j in range(len(soup.findAll("div",class_="zyjs-jjtj")[i].findAll("li"))):
        names.append(soup.findAll("div",class_="zyjs-jjtj")[i].findAll("li")[j].get_text())
        introduce_url.append(URL_PREF+soup.findAll("div",class_="zyjs-jjtj")[i].findAll("li")[j].a["href"].split("/")[-1])
        department.append(soup.findAll("div",class_="zyjs-jjtj")[i].find("div",class_="jjtj").get_text())
```

```python
data=pd.DataFrame(np.zeros(len(names)*3).reshape(len(names),3),columns=["姓名","个人介绍链接","所在院系"])
data["姓名"]=names
data["个人介绍链接"]=introduce_url
data["所在院系"]=department
data.head()
```

```python
def Spider():
    headers={"User-Agent": "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.120 Safari/537.36 Core/1.77.119.400 QQBrowser/10.9.4817.400"}
    URL = 'http://ds.zufe.edu.cn/szdw2/zyjs1.htm'#整个网页链接
    URL_PREF="http://ds.zufe.edu.cn/info/1092/"#每个老师的个人链接
    res=requests.get(URL,headers=headers)
    res.encoding="utf-8"
    html_source=res.text
    status_code=res.status_code
    #print(status_code)
    soup=BeautifulSoup(html_source,"html.parser")
    names=[]
    introduce_url=[]
    department=[]
    row=len([j for i in soup.findAll("div",class_="zyjs-jjtj") for j in i.findAll("li")])
    for i in range(len(soup.findAll("div",class_="zyjs-jjtj"))):
        for j in range(len(soup.findAll("div",class_="zyjs-jjtj")[i].findAll("li"))):
            names.append(soup.findAll("div",class_="zyjs-jjtj")[i].findAll("li")[j].get_text())
            introduce_url.append(URL_PREF+soup.findAll("div",class_="zyjs-jjtj")[i].findAll("li")[j].a["href"].split("/")[-1])
            department.append(soup.findAll("div",class_="zyjs-jjtj")[i].find("div",class_="jjtj").get_text())
     
    data=pd.DataFrame(np.zeros(row*3).reshape(row,3),columns=["姓名","个人介绍链接","所在院系"])
    data["姓名"]=names
    data["个人介绍链接"]=introduce_url
    data["所在院系"]=department
    
    return data
```
