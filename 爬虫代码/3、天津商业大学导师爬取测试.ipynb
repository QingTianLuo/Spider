```python
import os,sys    
import re    
import requests    
from bs4 import  BeautifulSoup    
import traceback    
import pandas as pd  
import numpy as np
import copy
```

```python
headers={  
'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.120 Safari/537.36 Core/1.77.119.400 QQBrowser/10.9.4817.400'  
} 
```

```python
res=requests.get(url,headers=headers)
res.encoding="utf-8"
status_code=res.status_code
html_source=res.text
print(status_code)
soup=BeautifulSoup(html_source,"html.parser")
department={}
for i in soup.findAll("a",class_="menu0_1_")[14:22]:
    department[i.get_text().split("\xa0")[1]]=URL+i["href"]
```

```python
Data=pd.DataFrame(np.zeros(1*3).reshape(1,3),columns=["姓名","个人链接","所在院系"])
for j in department:
    res1=requests.get(department[j],headers=headers)
    res1.encoding="utf-8"
    #print(res1.status_code)
    soup1=BeautifulSoup(res1.text,"html.parser")
    names=[]
    Personal_links=[]
    for i in soup1.findAll("ul",class_="newslist")[0].findAll("li"):
        Personal_links.append(URL+i.findAll("a")[1]["href"][2:])
        names.append(i.findAll("a")[1]["title"])
    row=len(soup1.findAll("ul",class_="newslist")[0].findAll("li"))
    data=pd.DataFrame(np.zeros(row*3).reshape(row,3),columns=["姓名","个人链接","所在院系"])
    data["姓名"]=names
    data["个人链接"]=Personal_links
    data["所在院系"]=j
    Data=Data.append(data,ignore_index=True)
Data.drop(Data.index[0], inplace=True)
Data = Data.reset_index(drop=True)
```

```python
Data1=copy.deepcopy(Data)
```

```python
Data=Data.drop([31,32,48,63,82])
```

```python
Data["个人介绍"]=0
Data["图片"]=0
Data["联系方式"]=0
```

```python
for i in range(len(Data)):
    res=requests.get(Data.iloc[i,1],headers=headers)
    res.encoding="utf-8"
    #print(res.status_code)
    soup=BeautifulSoup(res.text,"html.parser")
    row=len(soup.findAll("div",class_="v_news_content")[0].findAll("p"))
    Data.iloc[i,4]=URL+soup.findAll("div",class_="v_news_content")[0].findAll("p")[0].find("img",class_="img_vsb_content")["orisrc"]
    Data.iloc[i,5]=soup.findAll("div",class_="v_news_content")[0].findAll("p")[-2].get_text().split("：")[-1]
    num=""
    for j in range(1,row-2):
        num+=soup.findAll("div",class_="v_news_content")[0].findAll("p")[j].get_text().replace("\xa0","")
    Data.iloc[i,3]=num
```

```python
data=pd.merge(Data1.iloc[:,0],Data,on="姓名",how="left")
```
